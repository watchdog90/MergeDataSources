```python
import pandas as pd
pd.set_option('expand_frame_repr', False) 

pd_display_cols = 50
pd_display_rows = 20
pd.set_option('display.max_rows', pd_display_rows)
pd.set_option('display.min_rows', pd_display_rows)
```

# Load BIM_data - 10 columns

```python
df_BIM_data = pd.read_csv('BIM_data.csv')
df_BIM_data
```

# Load economic data - 4 columns

```python
df_economic_data = pd.read_csv('economic_data.csv')
df_economic_data
```

# Load Sustainable data - 4 columns

```python
df_sustainable_data = pd.read_csv('sustainable_data.csv')
df_sustainable_data
```

# Merge data from 3 sources - 18 columns

```python
df1 = pd.merge(df_BIM_data, df_economic_data, left_index=True, right_index=True)
df1
```

```python
df = pd.merge(df1, df_sustainable_data, left_index=True, right_index=True)
df
```
